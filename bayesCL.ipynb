# Clasificador 

```python
import csv
import numpy as np
import pandas as pd
import sklearn
from sklearn.preprocessing import Imputer
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
```

# Lectura del archivo

```python
datos = pd.read_csv('bank-additional-full.csv', sep=';')
#print(datos['y'])

datosPr = datos
datosPr.describe(include= 'all')
```

```python
#Identificar valores perdidos
for value in ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact',
             'month', 'day_of_week','campaign', 'poutcome', ]:
    print (value,":", sum(datosPr[value] == '?'))
```

# Procesamiento de datos

```python
#transformar los datos categoricos
le = preprocessing.LabelEncoder()
job_cat = le.fit_transform(datos.job)
marital_cat = le.fit_transform(datos.marital)
education_cat   = le.fit_transform(datos.education)
default_cat = le.fit_transform(datos.default)
housing_cat = le.fit_transform(datos.housing)
loan_cat = le.fit_transform(datos.loan)
contact_cat = le.fit_transform(datos.contact)
month_cat = le.fit_transform(datos.month)
day_of_week_cat = le.fit_transform(datos.day_of_week)
campaign_cat = le.fit_transform(datos.campaign)
poutcome_cat = le.fit_transform(datos.poutcome)

#initialize the encoded categorical columns
datosPr['job_cat'] = job_cat
datosPr['marital_cat'] = marital_cat
datosPr['education_cat'] = education_cat
datosPr['default_cat'] = default_cat
datosPr['housing_cat'] = housing_cat
datosPr['loan_cat'] = loan_cat
datosPr['contact_cat'] = contact_cat
datosPr['month_cat'] = month_cat
datosPr['day_of_week_cat'] = day_of_week_cat
datosPr['campaign_cat'] = campaign_cat
datosPr['poutcome_cat'] = poutcome_cat

#Eliminar las columnas orginales
borrar_columnas = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact',
             'month', 'day_of_week','campaign', 'poutcome', ]

datosPr = datosPr.drop(borrar_columnas, axis = 1)
#print(datosPr)
```

```python
#Organizar las columnas
datosPr = datosPr.reindex_axis(['age', 'job_cat', 'marital_cat', 'education_cat',
                                    'default_cat', 'housing_cat', 'loan_cat',
                                    'contact_cat', 'month_cat', 'day_of_week_cat', 'duration',
                                    'campaign_cat', 'pdays', 'previous', 
                                    'poutcome_cat', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m',
                               'nr.employed', 'y'], axis= 1)
```

# Estandarizar datos

```python
columnas = ['age', 'job_cat', 'marital_cat', 'education_cat',
                                    'default_cat', 'housing_cat', 'loan_cat',
                                    'contact_cat', 'month_cat', 'day_of_week_cat', 'duration',
                                    'campaign_cat', 'pdays', 'previous', 
                                    'poutcome_cat', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m',
                               'nr.employed']

scaled_features = {}
for i in columnas:
    mean, std = datosPr[i].mean(), datosPr[i].std()
    scaled_features[i] = [mean, std]
    datosPr.loc[:, i] = (datosPr[i] - mean)/std
```

# Division de datos para el entrenamiento y pruebas

```python
caracteristicas = datosPr.values[:,:20]
objetivo = datosPr.values[:,20]

caracteristicas_train, caracteristicas_test,objetivo_train, objetivo_test = train_test_split(caracteristicas,
                                                                                             objetivo, test_size = 0.33, 
                                                                                             random_state = 10)
clasificador = GaussianNB()
clasificador.fit(caracteristicas_train, objetivo_train)
obj_pred = clasificador.predict(caracteristicas_test)

#Se muestra la precision del modelo y la matriz de confusion

print ('Precision: ', accuracy_score(objetivo_test, obj_pred, normalize = True))
print ('Matriz de Confusion\n', confusion_matrix(objetivo_test, obj_pred))
print ('Filas: Valor Real\nColumnas: Valor Predicho')
```
